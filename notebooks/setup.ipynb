# Setup & Data Check
This notebook verifies the environment, loads sample MRI slices, and performs a corruption demo.

```python
import os, glob
from PIL import Image
import matplotlib.pyplot as plt
from src.medrestor_gan.dataset import pil_to_tensor_gray, tensor_to_pil_gray, add_noise, random_mask
paths = sorted(glob.glob('../data/raw/*.jpg'))[:4]
imgs = [Image.open(p).convert('L').resize((256,256)) for p in paths]
import torch
fig, axs = plt.subplots(len(imgs), 3, figsize=(9, 3*len(imgs)))
for i, im in enumerate(imgs):
    t = pil_to_tensor_gray(im)
    noisy = add_noise(t)
    m = torch.from_numpy(random_mask(256,256)).unsqueeze(0)
    axs[i,0].imshow(im, cmap='gray'); axs[i,0].set_title('Clean'); axs[i,0].axis('off')
    axs[i,1].imshow((noisy*m).squeeze(), cmap='gray'); axs[i,1].set_title('Corrupted'); axs[i,1].axis('off')
    axs[i,2].imshow(m.squeeze(), cmap='gray'); axs[i,2].set_title('Mask'); axs[i,2].axis('off')
plt.tight_layout(); plt.show()
```
